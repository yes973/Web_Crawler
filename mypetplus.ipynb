마이펫플러스(https://www.mypetplus.co.kr/)에 입점한 동물병원, 상품, 소재지를 크롤링합니다.
해당 사이트는 동적 웹페이지로 구축되었기 때문에 Selenium webdriver를 사용했으며
완전한 데이터를 불러오려면 (페이지 업데이트 상황에 따라) 소스코드를 수정해야 할 수 있습니다.

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from numpy import nan as NA
from pandas import Series,DataFrame
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome('C:\chromedriver') #chromedriver의 위치를 지정합니다.
driver.get('https://www.mypetplus.co.kr/Product/ProductList/List?ctg1=26')

loc = []
item = []
hosp = []

css_selector_all = ['#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li.ng-scope.active > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(2) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(3) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(4) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(5) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(6) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(7) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(8) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(9) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth1 > li:nth-child(2) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(2) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(3) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth1 > li:nth-child(3) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(2) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(3) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(4) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(5) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(6) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(7) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(8) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(9) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(10) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(11) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(12) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth2.ng-scope > li:nth-child(9) > a'
                    ,'#container > div > div.sub-cate.ng-scope > ul.depth1 > li:nth-child(4) > a']

for objs in css_selector_all :
    time.sleep(3)
    driver.find_element_by_css_selector(objs).click()
    time.sleep(3)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    for codes in soup.find_all('i', class_='loc') :
        loc.append(codes.text)

    for codes in soup.find_all('p', class_='tit') :
        item.append(codes.text.split('-')[0].strip())
        hosp.append(codes.text.split('-')[1].strip())
    
    print('크롤링 개수 : ' + str(len(soup.find_all('i', class_='loc'))))

driver.close()
```

```python
mypet_hospital_list = pd.DataFrame(
    {'loc': loc,
     'hosp': hosp,
     'item': item 
    })
```
