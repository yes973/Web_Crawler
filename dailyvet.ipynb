소통하는 수의사신문 데일리벳(http://dailyvet.co.kr)의 기사와 본문을 추출하는 크롤러입니다.

```python
import datetime
import requests
import urllib.request
import urllib.error
import urllib.parse
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from numpy import nan as NA
from pandas import Series,DataFrame
```

```python
url_list=[]
date_list=[]
for page_num in range(1,100) :
    response = urllib.request.urlopen('http://www.dailyvet.co.kr/category/news/page/'+str(page_num))
    response_body = response.read()
    soup = BeautifulSoup(response_body.decode('utf-8'), 'html.parser')
    crawling_urls = soup.find_all('dt')[0:15]
    for url in crawling_urls :
        url_list.append(url.find('a')['href'])
    crawling_dates = soup.find_all('dd','date')
    for dates in crawling_dates :
        date_list.append(dates.text)

df = pd.DataFrame({'url' : url_list, 'postdate' : date_list})
df.postdate = pd.to_datetime(df.postdate,format='%Y-%m-%d')
```

먼저 메인페이지에서 최근 100개 페이지의 기사 URL과 등록일을 추출합니다.
이후 아래 예시와 같이 수집된 URL을 돌며 기사의 본문을 추출할 수 있습니다.
예시에서는 df 데이터프레임의 indexing을 통해 2019년 2월에 해당하는 기사만을 추출하고 있습니다.

```python
# 수집된 url에서 기사 본문을 추출하는 크롤러
#2월만 추출하고 싶은 경우
article_text = []
for url in df.loc[df.postdate.dt.month == 2].url :
    response = urllib.request.urlopen(url)
    response_body = response.read()
    soup = BeautifulSoup(response_body.decode('utf-8'), 'html.parser')
    article_body = soup.find_all('p', class_=None)[0:-2]
    for texts in article_body :
        article_text.append(texts.text)
```
