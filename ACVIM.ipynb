특정 기간 동안 수의내과학저널(https://onlinelibrary.wiley.com/journal/19391676)에 발표된 논문들의
제목, 초록, 저자, 발표일 등을 크롤링합니다. 

```python
import re
import json
import math
import datetime
import requests
import urllib.request
import urllib.error
import urllib.parse
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from numpy import nan as NA
from pandas import Series,DataFrame
```

```python
start="2009-07-21"
end="2019-07-21"

title=[]
contents=[]
description=[]
creator=[]
date=[]
section=[]

for loop_num in range(len(pd.date_range(start=start, end=end, freq='W'))-2) :
    loop_start=pd.date_range(start=start, end=end, freq='W')[loop_num].strftime('%Y%m%d')
    loop_end=pd.date_range(start=start, end=end, freq='W')[loop_num+1].strftime('%Y%m%d')
    response = urllib.request.urlopen('https://onlinelibrary.wiley.com/action/showFeed?ui=0&mi=n90onq&type=search&feed=rss&query=%2526Ppub%253D%25255B'+str(loop_start)+'%252BTO%252B'+str(loop_end)+'%25255D%2526SeriesKey%253D19391676%2526content%253DarticlesChapters%2526sortBy%253DEarliest%2526target%253Ddefault')
    response_body = response.read()
    soup = BeautifulSoup(response_body.decode('utf-8'), 'lxml')

    for items in soup.find_all('item') :
        title.append(items.find('title').string)
        contents.append(items.find('dc:description').string)
        description.append(items.find('description').string)
        creator.append(items.find('dc:creator').string)
        date.append(items.find('dc:date').string)
        section.append(items.find('prism:section').string)
```

```python
test = pd.DataFrame({'title' : title, 'contents' : contents,'description' : description, 'creator' : creator, 'date' : date, 'section' : section})
```
